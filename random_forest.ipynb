```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from scipy.stats import randint
```

```python
full_df = pd.read_csv("task/labeled_and_skeleton.csv")
meta_cols = ['gender','age','fairNumber']
```

```python
affect_cols = ['sessionId', 'timestamp', 'gender', 'age', 
       'affect', 'watchId', 'duration', 'fairNumber']
affect_df = full_df[full_df.affect.notna()][affect_cols]
affect_dummies_df = pd.get_dummies(affect_df, columns=meta_cols)
```

```python
context_cols = ['sessionId', 'timestamp', 'gender', 'age', 
       'context', 'watchId', 'duration', 'fairNumber']
context_df = full_df[full_df.context.notna()][context_cols]
context_dummies_df = pd.get_dummies(context_df, columns=meta_cols)
```

```python
X = affect_dummies_df.drop(['affect',"watchId"], axis=1)
y = affect_dummies_df['affect']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

param_dist = {'n_estimators': randint(50,500),
              'max_depth': randint(1,20)}


y_pred = rf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
X = context_dummies_df.drop(['context',"watchId"], axis=1)
y = context_dummies_df['context']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

param_dist = {'n_estimators': randint(50,500),
              'max_depth': randint(1,20)}


y_pred = rf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
